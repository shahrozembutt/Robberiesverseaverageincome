### Postcode, Latitude, and Longitude

-It reads the Geospatial_Coordinates and outputs the postal code, along with the longitude and the latitutide. 

```python
df1 = pd.read_csv('Geospatial_Coordinates.csv')
df1
```

### Using the Wikipedia Page to create Data

-Using the Wikipedia page provided to create a dataset while assuring that values aren't being duplicated but rather removed if there is a duplicate available.

```python
source = requests.get('https://en.wikipedia.org/w/index.php?title=List_of_postal_codes_of_Canada:_M&oldid=945633050').text
soup = BeautifulSoup(source,'lxml')
table = str(soup.table)
dfs = pd.read_html(table)
df = dfs[0]

# Removing the rows where Neighbourhood is 'Not assigned'
df0 = df[df.Neighbourhood != 'Not assigned']

#Drop neighbourhood names with same Postcode
df2 = df0.drop_duplicates(subset=['Postcode'])
df2.reset_index(inplace=True)
df2 = df2.drop('index', 1)
df2
```

### Datasets being Merged

-After having the Geospatial_coordinates we combine both the tables that we had created along with the postal codes and created the table below.

```python
df3 = pd.merge(df1,df2,on = 'Postcode')
df3
```

### Merging the Neighbourhood Crime Rate Dataset

-They called for the csv link for the Neighbourhood Crime Rate and merging it with the table. 

```python
#calling for the file neighbourhood crime rate to then analyze and read the dataset
df4 = pd.read_csv('Neighbourhood_Crime_Rates.csv')
#we had to merge the dataset after the continuation with the neighbourhood
df5 = pd.merge(df3,df4,on = 'Neighbourhood')
df5
```

### Neighbourhood Data

-It provides detailed information about each neighborhood within Toronto. 

```python
# it calls and reads for the neighbourhood profile
df6 = pd.read_csv('neighbourhood-profiles.csv')
#we had to merge the dataset after the continuation with the neighbourhood
df7 = pd.merge(df6,df5,on = 'Neighbourhood')
df7['Total - Income statistics in 2015 for the population aged 15 years and over in private households'] = pd.to_numeric(df7['Total - Income statistics in 2015 for the population aged 15 years and over in private households'], errors='coerce')
df7
```

### It visualizes all the Neighbourhoods of the above data frame using Folium

```python
#it provides the latitude and longitude the map starts out with and the zoom would be at 10
map_toronto = folium.Map(location=[43.65,-79.34],zoom_start=10)

#Postal Code Markers
for lat, lng, neighbourhood in zip(df7['Latitude'], df7['Longitude'],df7['Neighbourhood']):
    label = '{}'.format(neighbourhood)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=10,
        popup=label,
        color='black',
        fill=True,
        fill_color="#3186cc",
        ).add_to(map_toronto)  
map_toronto  

title_html = '''
             <h3 align="center" style="font-size:14px"><b>{}</b></h3>
             '''.format('ROBBERY LIKELYHOOD: BLUE - very unlikely, GREEN - unlikely, YELLOW - maybe, ORANGE - likely, RED - very likely') 
map_toronto.get_root().html.add_child(folium.Element(title_html))

#it was based on robberies that occured overall
for lat, lng, neighbourhood, robbery in zip(df7['Latitude'], df7['Longitude'],df7['Neighbourhood'],df7['Robbery_2015']):
    #less than 10 robberies creates a red dot representing it's very unlikely to be robbed
    if (robbery <= 10):
        label = '{}'.format(neighbourhood)
        label = folium.Popup(label, parse_html=True)
        folium.CircleMarker(
            [lat, lng],
            radius=15,
            popup=label,
            color='blue',
            fill=True,
            fill_color="blue",
            ).add_to(map_toronto)
    #from 11 to 40 robberies creates a green dot representing it's unlikely to be robbed        
    if (11 < robbery <= 40):
        label = '{}'.format(neighbourhood)
        label = folium.Popup(label, parse_html=True)
        folium.CircleMarker(
            [lat, lng],
            radius=15,
            popup=label,
            color='green',
            fill=True,
            fill_color="green",
            ).add_to(map_toronto)  
    #from 41 to 60 robberies creates a yellow dot representing it's maybe to be robbed        
    if (41 < robbery <= 60):
        label = '{}'.format(neighbourhood)
        label = folium.Popup(label, parse_html=True)
        folium.CircleMarker(
            [lat, lng],
            radius=15,
            popup=label,
            color='yellow',
            fill=True,
            fill_color="yellow",
            ).add_to(map_toronto)  
    #from 61 to 90 robberies creates a blue dot representing it's likely to be robbed        
    if (61 < robbery <= 90):
        label = '{}'.format(neighbourhood)
        label = folium.Popup(label, parse_html=True)
        folium.CircleMarker(
            [lat, lng],
            radius=15,
            popup=label,
            color='orange',
            fill=True,
            fill_color="orange",
            ).add_to(map_toronto)  
    #from 90+ robberies creates a black dot representing it's very likely to be robbed        
    if (robbery > 91):
        label = '{}'.format(neighbourhood)
        label = folium.Popup(label, parse_html=True)
        folium.CircleMarker(
            [lat, lng],
            radius=15,
            popup=label,
            color='red',
            fill=True,
            fill_color="red",
            ).add_to(map_toronto)  
map_toronto
```
